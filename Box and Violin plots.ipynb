```python
import seaborn as sns
import numpy as np
```

```python
tips = sns.load_dataset('tips')
```

```python
tips
```

```python
sns.boxplot(x='day',y='tip',data=tips,hue='smoker')
#box shows the max range of values
#upper line is max,lower line is min
#dots are outliers
```

```python
sns.boxplot(x='day',y='tip',data=tips,hue='time')
```

```python
sns.boxplot(x='day',y='tip',data=tips,hue='sex')
```

```python
sns.boxplot(x='day',y='total_bill',data=tips,hue='smoker')
```

```python
sns.boxplot(x='day',y='total_bill',data=tips,hue='sex')
```

```python
sns.boxplot(x='day',y='total_bill',data=tips,hue='time')
```

```python
sns.boxplot(x='time',y='tip',data=tips,hue='smoker')
```

```python
sns.boxplot(x='time',y='total_bill',data=tips,hue='smoker')
```

```python
#darker line shows where mostof the values are concentrated
# dot in darker line is the avg
#Highest tip of violin shows max value in data
sns.violinplot(x='time',y='total_bill',data=tips,hue='smoker')
```

```python
sns.violinplot(x='day',y='total_bill',data=tips,hue='sex')
```

```python
sns.violinplot(x='time',y='total_bill',data=tips,hue='sex',split =True,palette='Set3')
```

```python
sns.violinplot(x='day',y='tip',data=tips,hue='smoker')
```

```python
sns.violinplot(x='day',y='tip',data=tips,hue='sex',split=True)
```
